```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('adc_teraterm_out_log.txt', 'r') as file:
    # Read lines from the file and split each line into hexadecimal numbers
    lines = [line.strip().split() for line in file.readlines()]

    decimal_list = [[int(num, 16) for num in line] for line in lines]
    
decimal_list
```

```python
flat_list = [item for sublist in decimal_list for item in sublist]
flat_list
```

```python
hist = plt.hist(flat_list, bins = 500, edgecolor='black')
plt.xlabel( "keV")
plt.ylabel("counts")
plt.title("Barium Spectra")

plt.xlim(0, 200 ) # Set the limits for the y-axis
plt.show()
#print(,"the peak value in the dataset is at" max(flat_list))
```
